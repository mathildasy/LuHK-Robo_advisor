```python
df.iloc[::-1]
df.fillna(method='bfill', axis='index', inplace=True)
df
```

```python
assets = df.columns[1:]
assets
```

```python
ll = df[assets[2]].tolist() 
ll = [float(l.replace(',', '')) for l in ll]
df[assets[2]] = ll
```
